```
def return_level(xi, beta, p):  # Fixed mathematical formula for GPD return levels
    if beta <= 0.0:
        raise ValueError('Rate parameter beta must be positive')
    return beta * (1 - (-np.log(1 - p))**(-1 / xi))  # Improved formula
```

```
def gpd_nll_score(data, xi, beta):  # Added input validation and error handling
    if beta <= 0:
        raise ValueError('Beta must be greater than 0')
    if xi <= 0:
        raise ValueError('Xi must be greater than 0')
    import warnings
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', category=RuntimeWarning)
        return -np.sum(np.log(gpd.pdf(data, xi, beta)))  # Prevent log domain errors
```

```
def gpd_exceedance_quantile(q, xi, beta):  # Added input validation with bounds checking
    if not (0 < q < 1):
        raise ValueError('Quantile must be between 0 and 1')
    return beta * (1 - (-np.log(q))**(-1 / xi))  # Adjusted formula for quantile
```

```
def preprocess_fire_days(data):  # Added logging for data loss
    original_length = len(data)
    data = data.dropna()  # Data prep
    if len(data) < original_length:
        logging.warning(f'Data loss occurred: {original_length - len(data)} entries lost')  # Log message
```

```
def gpd_objective_fixed_xi(data, xi, beta):  # Improved numerical stability
    return np.sum((data - (xi + beta))**2)  # Adjusted calculation for stability
```
